EXEMPLO BASE PARA CONSULTA DO JSON E ARMAZENAMENTO LOCAL

```python
# Bibliotecas 

import os
import requests
import json
```

----------------- Consulta Cliente --------------------------

```python
# Define o caminho da pasta
folder_path = "C:\\NOMEEMPRESA\\cliente"

# Verifica se a pasta existe, se não, cria
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

url = "https://NOMEEMPRESA.varejofacil.com/api/v1/pessoa/clientes"
headers = {
    "Accept": "application/json",
    "x-api-key": "TOKEN"
}

# Faz a primeira solicitação para obter o total de registros
params = {
    "start": 0,
    "count": 1  # Apenas para obter o total de registros sem muita sobrecarga
}
response = requests.get(url, headers=headers, params=params)

if response.status_code == 200:
    data = response.json()
    total_records = data.get("total", 0)
    print("Total de registros:", total_records)

    # Calcula o número de solicitações necessárias
    num_requests = total_records // 100 + (1 if total_records % 100 != 0 else 0)

    # Faz as solicitações em loop
    for i in range(num_requests):
        start = i * 100
        params = {
            "start": start,
            "count": 100
        }
        response = requests.get(url, headers=headers, params=params)

        if response.status_code == 200:
            # Salva o JSON em um arquivo
            file_path = os.path.join(folder_path, f"dados_{start}-{start+99}.json")
            with open(file_path, "w") as file:
                json.dump(response.json(), file)
            print(f"JSON salvo com sucesso em: {file_path}")
        else:
            print(f"Falha na requisição {i+1}:", response.status_code)

else:
    print("Falha na requisição:", response.status_code)
```
