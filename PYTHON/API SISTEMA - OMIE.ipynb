EXEMPLO BASE PARA CONSULTA DO JSON E ARMAZENAMENTO LOCAL

```python
import os
import requests
import json

# Definir os parâmetros da API
url = "https://app.omie.com.br/api/v1/financas/mf/"
app_key = " "  # Substitua pelo seu OMIE_APP_KEY
app_secret = "  "  # Substitua pelo seu OMIE_APP_SECRET
call = "ListarMovimentos"

# Definir o número da página inicial e o total de registros por página
pagina = 1
registros_por_pagina = 1000

# Lista para armazenar todos os registros
todos_registros = []

# Caminho do diretório onde os arquivos serão salvos
diretorio = r"C:\Api_Cliente"

# Criar o diretório se ele não existir
os.makedirs(diretorio, exist_ok=True)

# Loop para buscar todas as páginas e armazenar os registros
while True:
    # Configurar o payload da requisição
    payload = {
        "call": call,
        "app_key": app_key,
        "app_secret": app_secret,
        "param": [
            {
                "nPagina": pagina,
                "nRegPorPagina": registros_por_pagina
            }
        ]
    }

    # Fazer a requisição POST
    response = requests.post(url, json=payload)
    data = response.json()

    # Verificar se há registros retornados
    registros = data.get('movimentos', [])
    if not registros:
        break

    # Adicionar os registros à lista de todos registros
    todos_registros.extend(registros)

    print(f"Página {pagina} processada com sucesso.")

    # Avançar para a próxima página
    pagina += 1

# Caminho do arquivo único onde todos os registros serão salvos
caminho_arquivo_unico = os.path.join(diretorio, "Financas.json")

# Salvar todos os registros no arquivo JSON único
with open(caminho_arquivo_unico, 'w', encoding='utf-8') as f:
    json.dump(todos_registros, f, ensure_ascii=False, indent=4)

print(f"Todos os registros foram salvos em {caminho_arquivo_unico}")
```
