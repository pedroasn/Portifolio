```python
import requests
import json
import os
from datetime import datetime, timedelta
```

**CRM BONUS GERADO**

```python
# URL e cabeçalhos
url = "https://api.crmbonus.com/bonusGerados"

headers = {
  'Authorization': " ",
  'Token': " ",
  'accept': 'application/json',
  'content-type': 'application/json'
}

# Diretório onde os JSONs serão salvos
output_directory = " "  # Inclua a barra final
os.makedirs(output_directory, exist_ok=True)

# Defina o intervalo de anos que deseja buscar (exemplo: 2023 e 2024)
ano_atual = datetime.now().year
anos = [ano_atual]

# Loop pelos anos e meses
for ano in anos:
    for mes in range(1, 13):  # Meses de 1 a 12
        # Gerar as datas de início e fim do mês
        data_inicio = datetime(ano, mes, 1)
        ultimo_dia = (data_inicio.replace(month=mes % 12 + 1, day=1) - timedelta(days=1)).day
        data_fim = datetime(ano, mes, ultimo_dia)
        
        # Formatar o período no formato requerido pela API
        periodo = f"{data_inicio.strftime('%Y-%m-%dT00:00:00')} TO {data_fim.strftime('%Y-%m-%dT23:59:59')}"
        payload = json.dumps({"w_data": periodo})
        
        # Fazer a requisição
        response = requests.request("GET", url, headers=headers, data=payload)
        
        # Nome do arquivo para o mês e ano atual
        output_filename = f"Bonus_Gerado_{ano}_{mes:02}.json"
        output_path = os.path.join(output_directory, output_filename)
        
        # Verificar e salvar a resposta
        if response.status_code == 200:
            with open(output_path, "w", encoding="utf-8") as json_file:
                json.dump(response.json(), json_file, indent=4, ensure_ascii=False)
            print(f"Arquivo salvo com sucesso: {output_path}")
        else:
            print(f"Erro {response.status_code} ao buscar dados para {ano}-{mes:02}: {response.text}")
```
